```python
import matplotlib.pyplot as plt
import numpy as np
from math import comb
from scipy.stats import binom

# Parameters
n = 10  # number of trials
p = 0.5  # probability of success
x = np.arange(0, n+1)  # possible outcomes (0 to n)

# Binomial coefficient (combination)
binom_coeff = [comb(n, k) for k in x]

# Success term: p^k
success_term = [p**k for k in x]

# Failure term: (1-p)^(n-k)
failure_term = [(1-p)**(n-k) for k in x]

# Full binomial PMF
binomial_pmf = binom.pmf(x, n, p)

# Plotting
fig, axs = plt.subplots(4, 1, figsize=(8, 12))

# Plot binomial coefficient
axs[0].bar(x, binom_coeff, color='lightcoral')
axs[0].set_title(r'$\binom{n}{k}$ (Binomial Coefficient)')
axs[0].set_xlabel('k (Number of Successes)')
axs[0].set_ylabel('Value')

# Plot success term p^k
axs[1].bar(x, success_term, color='lightblue')
axs[1].set_title(r'$p^k$ (Success Term)')
axs[1].set_xlabel('k (Number of Successes)')
axs[1].set_ylabel('Value')

# Plot failure term (1-p)^(n-k)
axs[2].bar(x, failure_term, color='lightgreen')
axs[2].set_title(r'$(1-p)^{n-k}$ (Failure Term)')
axs[2].set_xlabel('k (Number of Successes)')
axs[2].set_ylabel('Value')

# Plot full binomial distribution (PMF)
axs[3].bar(x, binomial_pmf, color='purple')
axs[3].set_title('Binomial Distribution PMF')
axs[3].set_xlabel('k (Number of Successes)')
axs[3].set_ylabel('Probability')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.stats import bernoulli
import numpy as np

# Parameters for Bernoulli distribution
p = 0.6  # probability of success
size = 1000  # number of experiments

# Generate Bernoulli data
data_bern = bernoulli.rvs(size=size, p=p)

# Plotting the Bernoulli distribution
fig, ax = plt.subplots(figsize=(8, 6))

# Using seaborn to plot the distribution
ax = sb.histplot(data_bern, kde=False, color='crimson', bins=2, stat="density", discrete=True, edgecolor='black')

# Add annotations for Success and Failure
plt.text(0, 0.4, 'Failure (0)', fontsize=14, color='blue', horizontalalignment='center')
plt.text(1, 0.4, 'Success (1)', fontsize=14, color='green', horizontalalignment='center')

# Set labels
ax.set(xlabel='Outcome (0=Failure, 1=Success)', ylabel='Frequency', title=f'Bernoulli Distribution (p = {p})')

# Add grid
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()

# Additional Explanation (Success vs. Failure Probability)
success_count = np.sum(data_bern)
failure_count = size - success_count

print(f"Out of {size} trials, there were {success_count} successes and {failure_count} failures.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import comb

# Parameters for Bernoulli and Binomial
p = 0.6  # Probability of success
n_trials = 10  # Number of trials for Binomial

# Values for Bernoulli (x = 0, 1)
x_bern = np.array([0, 1])
bern_success = p**x_bern
bern_failure = (1 - p)**(1 - x_bern)
bern_pmf = bern_success * bern_failure

# Values for Binomial (k = 0 to n_trials)
k_binom = np.arange(0, n_trials + 1)
binom_coeff = np.array([comb(n_trials, k) for k in k_binom])
binom_success = p**k_binom
binom_failure = (1 - p)**(n_trials - k_binom)
binom_pmf = binom_coeff * binom_success * binom_failure

# Plotting Bernoulli equation components
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Bernoulli Equation Visualization
ax[0].bar(x_bern - 0.1, bern_success, width=0.2, label=r'$p^x$', color='lightblue')
ax[0].bar(x_bern + 0.1, bern_failure, width=0.2, label=r'$(1-p)^{1-x}$', color='lightgreen')
ax[0].bar(x_bern, bern_pmf, width=0.1, label='PMF', color='crimson')
ax[0].set_title('Bernoulli Distribution Components')
ax[0].set_xlabel('x (0=Failure, 1=Success)')
ax[0].set_ylabel('Value')
ax[0].legend()

# Binomial Equation Visualization
ax[1].bar(k_binom - 0.2, binom_coeff, width=0.2, label=r'$\binom{n}{k}$', color='lightcoral')
ax[1].bar(k_binom, binom_success, width=0.2, label=r'$p^k$', color='lightblue')
ax[1].bar(k_binom + 0.2, binom_failure, width=0.2, label=r'$(1-p)^{n-k}$', color='lightgreen')
ax[1].bar(k_binom, binom_pmf, width=0.1, label='PMF', color='purple')
ax[1].set_title(f'Binomial Distribution Components (n={n_trials})')
ax[1].set_xlabel('k (Number of Successes)')
ax[1].set_ylabel('Value')
ax[1].legend()

# Show plots
plt.tight_layout()
plt.show()
```

To visualize the **Poisson distribution** with animated graphs in Python, we can use **matplotlib**'s animation features. The Poisson distribution models the number of events happening in a fixed interval of time or space, given a constant average rate of occurrence (\( \lambda \)).

We'll create an animated graph that changes the value of \( \lambda \) (the average rate), showing how the distribution evolves as \( \lambda \) increases.

### Steps:
1. We'll generate Poisson-distributed data for different values of \( \lambda \).
2. We'll animate the graph to show how the distribution changes as \( \lambda \) increases.
3. The graph will update in real-time to illustrate this evolution.

### Explanation of the Code:

1. **Libraries**:
   - `numpy` is used for numerical operations.
   - `scipy.stats.poisson` generates the Poisson probability mass function (PMF).
   - `matplotlib.pyplot` is used to plot the data, and `FuncAnimation` creates the animation.

2. **Parameters**:
   - \( x \) is the range of event counts (from 0 to 20).
   - \( \lambda \) changes dynamically from 1 to 15 in the animation.

3. **Animation**:
   - The `update` function recalculates the Poisson distribution's PMF for each value of \( \lambda \) and updates the graph.
   - `FuncAnimation` animates the graph by updating the plot every 300 milliseconds (`interval=300`).

4. **Result**:
   - The plot will animate, showing how the Poisson distribution changes as \( \lambda \) increases from 1 to 15. The title of the graph updates to reflect the current value of \( \lambda \).

You can run this code in a Python environment that supports graphical outputs, such as Jupyter Notebooks or any local environment with `matplotlib` installed. The animation will visualize how the shape of the Poisson distribution changes based on the average rate \( \lambda \).

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from matplotlib.animation import FuncAnimation

# Set up the figure and axis
fig, ax = plt.subplots()

# Initial settings for the plot
x = np.arange(0, 20)
ax.set_xlim(0, 20)
ax.set_ylim(0, 0.4)
line, = ax.plot([], [], lw=2)

# Initialize plot elements (empty)
def init():
    line.set_data([], [])
    return line,

# Function to update the plot for each frame in the animation
def update(frame):
    # Poisson distribution with mean `lambda = frame`
    pmf = poisson.pmf(x, frame)
    line.set_data(x, pmf)
    ax.set_title(f"Poisson Distribution ( = {frame})")
    return line,

# Create animation
anim = FuncAnimation(fig, update, frames=np.arange(1, 15, 0.5), init_func=init, blit=False, interval=300)
plt.ion()


# Show the animated plot
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
anim.save('poisson_distribution.gif', writer='imagemagick')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_value = 3  # average rate ()
k_values = np.arange(0, 15)  # range of k values

# Poisson probabilities for k values
poisson_probabilities = poisson.pmf(k_values, lambda_value)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(k_values, poisson_probabilities, color='skyblue', alpha=0.7)
plt.title(f'Poisson Distribution ( = {lambda_value})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability P(X = k)')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(poisson_probabilities) + 0.05)  # setting y limit for better view
plt.show()
```
