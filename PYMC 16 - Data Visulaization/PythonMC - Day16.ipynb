```python
import numpy as np
import matplotlib.pyplot as plt

def plot_bitwise_xor():
    # Define the range of x values
    x = np.arange(-10, 11, 1)  # Integer values from -10 to 10
    z = x ^ 2  # Bitwise XOR with 2
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.stem(x, z, basefmt=" ", label=r'$z = x \, \oplus \, 2$')  # Remove 'use_line_collection' argument
    plt.title("Bitwise XOR Operation: $z = x \, \oplus \, 2$")
    plt.xlabel("x")
    plt.ylabel("z")
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.show()

# Call the function to visualize
plot_bitwise_xor()
```

```python
def visualize_bitwise_xor(x, y=2):
    # Convert the integers to 8-bit binary representations
    x_bin = format(x, '08b')
    y_bin = format(y, '08b')
    result = x ^ y
    result_bin = format(result, '08b')
    
    # Display the binary values
    print(f"Bitwise XOR operation for x = {x} and y = {y}:")
    print(f"  x  = {x_bin} (binary for {x})")
    print(f"  y  = {y_bin} (binary for {y})")
    print(f"-------------------------- XOR")
    print(f"  z  = {result_bin} (result in binary = {result})")
    
    # Step-by-step XOR process visualization
    print("\nStep-by-step XOR operation:")
    for i in range(len(x_bin)):
        print(f"  {x_bin[i]} XOR {y_bin[i]} = {result_bin[i]}")

# Call the function to visualize
visualize_bitwise_xor(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_ip_classes_fixed():
    classes = {
        'Class A': {'range': '0.0.0.0 - 127.255.255.255', 'first_octet_bits': '0xxxxxxx', 'subnet_mask': '255.0.0.0'},
        'Class B': {'range': '128.0.0.0 - 191.255.255.255', 'first_octet_bits': '10xxxxxx', 'subnet_mask': '255.255.0.0'},
        'Class C': {'range': '192.0.0.0 - 223.255.255.255', 'first_octet_bits': '110xxxxx', 'subnet_mask': '255.255.255.0'},
        'Class D': {'range': '224.0.0.0 - 239.255.255.255', 'first_octet_bits': '1110xxxx', 'subnet_mask': 'N/A (Multicast)'},
        'Class E': {'range': '240.0.0.0 - 255.255.255.255', 'first_octet_bits': '1111xxxx', 'subnet_mask': 'N/A (Reserved)'}
    }

    fig, ax = plt.subplots(figsize=(12, 7))  # Increase the figure size for more space
    
    # Define colors for each class
    colors = ['blue', 'green', 'coral', 'yellow', 'gray']
    bar_positions = np.arange(len(classes))
    bar_height = 0.8  # Increase the bar height for more vertical space

    # Create bars for each class
    ax.barh(bar_positions, [1]*len(classes), color=colors, height=bar_height, edgecolor='black')

    # Add class details to the plot
    for i, (cls, details) in enumerate(classes.items()):
        ax.text(0.05, i, f"{cls}: {details['range']}", va='center', fontsize=12, fontweight='bold', ha='left')
        ax.text(1.1, i, f"First Octet: {details['first_octet_bits']}", va='center', fontsize=10, ha='left')
        ax.text(1.8, i, f"Subnet Mask: {details['subnet_mask']}", va='center', fontsize=10, ha='left')
    
    # Set limits, labels, and title
    ax.set_xlim(0, 3)  # Increase x-limit to make more room for text
    ax.set_ylim(-0.5, len(classes) - 0.5)
    ax.set_yticks(bar_positions)
    ax.set_yticklabels(list(classes.keys()))
    ax.set_title('Visualization of IP Address Classes', fontsize=16, pad=20)
    
    # Hide the x-axis
    ax.xaxis.set_visible(False)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Call the function to visualize
visualize_ip_classes_fixed()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_square_function():
    # Define the range of x values
    x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10
    z = x ** 2  # Calculate z = x^2
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, z, label=r'$z = x^2$', color='b')
    plt.title("Plot of $z = x^2$")
    plt.xlabel("x")
    plt.ylabel("z")
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.legend()
    plt.show()

# Call the function to visualize
plot_square_function()
```
